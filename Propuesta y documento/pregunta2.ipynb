```python
import pandas as pd
from pathlib import Path
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
terremotos = pd.read_csv('query_limpio.csv')
terremotos = terremotos.drop('level_0', axis=1)
terremotos = terremotos.drop('Unnamed: 0', axis=1)
terremotos = terremotos.drop('index', axis=1)
regiones = set(terremotos['region'])
terremotos.head()
```

```python
regiones.remove('Coquimbo')
terremotos = terremotos.sort_values(by='mag',ascending=False)
terremotos
```

Si tomamos en consideración todos los registros de nuerto DataFrame, los mayores terremotos por región son los siguientes.

```python
mayores_sismos = terremotos[terremotos['region'] == 'Coquimbo'].head(1)
for region in regiones:
    otro1 = (terremotos[terremotos['region'] == region].head(1))
    mayores_sismos = mayores_sismos.merge(otro1, how='outer')
mayores_sismos
```
