| Dato | Llave en df | Descripción |
| --- | --- | --- |
| Coordenadas | *latitude* y *longitude* | Corresponden a la ubicación geospacial del sismo |
| Profundidad | *depth* | Corresponden a la profundidad en que se registra el sismo |
| Fallo en profundidad | *depthError* | Falta de certeza en los registros de profundidad |
| Magnitud | *mag* | Magnitud registrada del sismo |
| Fallo en magnitud | *magError* | Posible fallo en la medicion de la magnitud |
| Tipo de magnitud | *magType* | Forma en que fue registrada la magnitud |
| Estaciones en el registro | *magNst* | Es la cantidad de estaciones utilizadas para registrar el terremoto |
| Estaciones usadas | *nst* | Cantidad de estaciones usadas para determinar la ubicación del sismo |
| Brecha entre estaciones | *gap* | Desde un punto de referencia, es la mayor brecha entre dos estaciones (en grados) |
| Distancia estación-terremoto | *dmin* | Distancia en grados desde el epicentro a la estación más cercana |
| RMS | *rms* | Utilizado para medir la magnitud de la señal que varía en el tiempo |
| ID | *id* | Identificador asociado al evento |
| Fallo en ubicación | *horizontalError* | Falta de certeza en la ubicación del sismo |
| Fecha | *date* | Fecha en que ocurrió el evento |
| Hora | *hour* | Hora en que ocurrió el evento |
| Zona | *zone* | Comuna/Ciudad en que ocurrió el evento |
| Región | *region* | Región en que ocurrió el evento |
